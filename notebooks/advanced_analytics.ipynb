Mutual Fund Analytics Capstone Project

## Value at Risk (VaR)

VaR estimates the maximum expected loss at a specified confidence level.

Confidence Level: 95%

Output File:
- data/processed/var_cvar_report.csv

## Conditional Value at Risk (CVaR)

CVaR measures the average loss beyond the VaR threshold.

Used to evaluate downside risk in mutual fund investments.

## Rolling Sharpe Ratio

Rolling Sharpe Ratio tracks risk-adjusted returns over time.

Output:
- reports/rolling_sharpe_chart.png

## Monte Carlo Simulation

Monte Carlo simulation projects future NAV paths using random sampling techniques.

Implemented in:
- scripts/monte_carlo.py

## Portfolio Optimization

Portfolio optimization identifies efficient fund allocations for maximizing returns at a given risk level.

Implemented in:
- scripts/portfolio_optimizer.py

## Recommendation Engine

A recommendation system suggests mutual funds based on performance and risk characteristics.

Implemented using:
- Sharpe Ratio
- Return %
- Risk Metrics
